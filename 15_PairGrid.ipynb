```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('iris')
df
```

```python
g = sns.PairGrid(df)
g = g.map(plt.scatter)
plt.show()
```

```python
g = sns.PairGrid(df)
g = g.map(plt.scatter)
g = g.map_offdiag(plt.scatter)
g = g.map_diag(plt.hist)
plt.show()
```

```python
g = sns.PairGrid(df, hue='species')
g = g.map(plt.scatter)
g = g.map_offdiag(plt.scatter)
g = g.map_diag(plt.hist)
plt.show()
```

```python
g = sns.PairGrid(df, hue='species')
g = g.map(plt.scatter)
g = g.map_offdiag(plt.scatter)
g = g.map_diag(plt.hist, histtype='step')
plt.show()
```
