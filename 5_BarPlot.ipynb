```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
br = pd.read_csv('data.csv')
br
```

```python
plt.figure(figsize=(7,5))
sns.barplot(x="Duration",y="Calories", data=br)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.barplot(y="Maxpulse", data=br)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.barplot(x="Duration",y="Calories", data=br, hue='Duration')
plt.show()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
plt.figure(figsize=(7,5))
sns.barplot(x="sepal_length",y="petal_length", data=iris, hue='species')
plt.show()
```
