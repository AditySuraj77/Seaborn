```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('iris')
df
```

```python
plt.figure(figsize=(7,5))
sns.displot(x='sepal_length', data=df, hue='species')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.displot(x='sepal_length', data=df, hue='species', kind='kde')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.displot(x='sepal_length', data=df, hue='species', kind='ecdf')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.displot(x='sepal_length', data=df, hue='species',kde=True)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.displot(x='sepal_length', y='sepal_width', data=df)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.displot(x='sepal_length', y='sepal_width', data=df, kind='kde')
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.displot(x='sepal_length', y='sepal_width', data=df, kind='kde', rug=True)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.displot(x='sepal_length', y='sepal_width', data=df, kind='kde', col='species')
plt.show()
```
