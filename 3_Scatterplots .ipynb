```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv('data.csv')
data.head()

```

```python
plt.figure(figsize=(7,5))

sns.scatterplot(x="Duration", y="Calories", data= data)
plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.scatterplot(x="Duration", y="Calories", data= data, color='r')
plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.scatterplot(x="Duration", y="Calories", data= data, color='g', marker='+')
plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.scatterplot(x="Duration", y="Calories", data= data, color='g', marker='o', hue='Maxpulse')
plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.scatterplot(x="Duration", y="Calories", data= data, color='g', marker='o', hue='Maxpulse')

plt.xlim([50,200])
plt.show()

```

```python
plt.figure(figsize=(7,5))

sns.scatterplot(x="Duration", y="Calories", data= data, color='g', marker='o', hue='Maxpulse', style='Pulse')
plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.scatterplot(x="Duration", y="Calories", data= data, color='g', marker='o', hue='Maxpulse', size='Calories')
plt.show()
```
