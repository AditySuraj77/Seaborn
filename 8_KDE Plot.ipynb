```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('iris')
df
```

```python
sns.kdeplot(data=df, x='sepal_length')
plt.show()
```

```python
sns.kdeplot(data=df, x='sepal_width')
plt.show()
```

```python
sns.kdeplot(data=df, x='petal_length')
plt.show()
```

```python
sns.kdeplot(data=df, x='petal_width')
plt.show()
```

```python
sns.kdeplot(data=df)
plt.show()
```

```python
sns.kdeplot(data=df, x='sepal_length', bw_adjust=0.4)
plt.show()
```

```python
sns.kdeplot(data=df, x='sepal_length',bw_adjust=5, cut=1)
plt.show()
```

```python
sns.kdeplot(data=df, x='sepal_length', hue='species')
plt.show()
```

```python
sns.kdeplot(data=df, x='sepal_length', hue='species', multiple='stack')
plt.show()
```

```python
sns.kdeplot(data=df, x='sepal_length', hue='species', cumulative=True)
plt.show()
```

```python
sns.kdeplot(data=df, x='sepal_length', hue='sepal_width')
plt.show()
```
