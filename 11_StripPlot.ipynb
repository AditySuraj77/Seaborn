```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('test.csv')
df
```

```python
sns.stripplot(df['Age'])
plt.show()
```

```python
plt.figure(figsize=(11,8))
sns.stripplot(x = df['Age'], y = df['Name'])
plt.show()
```

```python
plt.figure(figsize=(11,8))
sns.stripplot(x = df['Age'], y = df['Name'], linewidth=10, size=20)
plt.show()
```

```python
plt.figure(figsize=(11,8))
sns.stripplot(x = df['Age'], y = df['Name'], palette='Set1', hue= df['Name'])
plt.show()
```

```python

sns.stripplot(x = df['Age'], y = df['Name'], linewidth=1, size=10, marker='<', alpha = .50)
plt.show()
```

```python
plt.figure(figsize=(11,8))
sns.stripplot(x = df['Age'], y = df['Name'], palette='Set1', hue= df['Name'])

sns.violinplot(x = df['Age'], y = df['Name'], palette='Set1', hue= df['Name'])
plt.show()
```
