```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('Player.csv')
df
```

```python
sns.catplot(y='Pulse', kind='count', data=df)

plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.catplot(x='Duration', y='Calories',  kind='violin', data=df)

plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.catplot(x='Duration', y='Calories',  kind='violin', data=df, height=3, aspect=1)

plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.catplot(x='Duration', y='Calories', data=df, col='Pulse')

plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.catplot(x='Calories', kind='count', data=df, col='Pulse', col_wrap=2,height=4)

plt.show()
```
