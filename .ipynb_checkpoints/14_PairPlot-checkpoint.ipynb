```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('iris')
df
```

```python
sns.pairplot(df, hue="petal_width")
plt.show()
```

```python
sns.pairplot(df, hue='species')
plt.show()
```

```python
sns.pairplot(df, hue='species', vars=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
plt.show()
```

```python
sns.pairplot(df, hue='species', vars=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], markers='<')
plt.show()
```

```python
sns.pairplot(df, hue='species', vars=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'], kind="reg", height=8)
plt.show()
```
