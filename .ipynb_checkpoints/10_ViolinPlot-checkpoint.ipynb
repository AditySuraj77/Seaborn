```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('iris_iris.csv')
df
```

```python
sns.violinplot(x='sepal.length', data=df)
plt.show()
```

```python
sns.violinplot(x='sepal.length', y='variety', data=df)
plt.show()
```

```python
sns.violinplot(x='sepal.length', y='variety', hue='variety', data=df)
plt.show()
```

```python
sns.violinplot(data=df, x='sepal.length', y='variety', hue='variety')
plt.show()
```

```python
sns.violinplot(data=df, x='sepal.length', y='variety', hue='variety', bw=0.5)
plt.show()
```

```python
sns.violinplot(data=df, x='sepal.length', y='variety', hue='variety', inner='stick')
plt.show()
```
