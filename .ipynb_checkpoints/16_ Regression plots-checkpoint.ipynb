```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('iris')
df
```

```python
sns.regplot(x = df['sepal_width'], y =df['sepal_length'])
plt.show()
```

```python
sns.regplot(x = df['sepal_width'], y =df['sepal_length'], color='r')
plt.show()
```

```python
sns.regplot(x = df['sepal_width'], y =df['sepal_length'], color='g', marker='>')
plt.show()
```

```python
sns.regplot(x = df['sepal_width'], y =df['sepal_length'], color='r', ci=100 , truncate=False)
plt.show()
```

```python
sns.regplot(x = df['sepal_width'], y =df['sepal_length'], color='r', ci=100 , truncate=False, logx=True)
plt.show()
```

```python
sns.regplot(x = df['sepal_width'], y =df['sepal_length'], color='r', ci=100 , truncate=False, logx=True, 
            line_kws={"color":"b","lw":10})
plt.show()
```

# Lmplot

```python
sns.lmplot(x='sepal_length', y='sepal_width', data=df)
plt.show()
```

```python
sns.lmplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.show()
```
