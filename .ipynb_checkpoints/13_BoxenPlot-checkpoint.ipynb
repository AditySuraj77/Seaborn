```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset('iris')
df
```

```python
plt.figure(figsize=(8,7))
sns.boxenplot(y=df["sepal_length"])
plt.show()
```

```python
plt.figure(figsize=(8,7))
sns.boxenplot(x=df["sepal_length"], y=df["species"], palette='Set1')
plt.show()
```

```python
plt.figure(figsize=(8,7))
sns.boxenplot(x=df["sepal_length"], y=df["species"], palette='spring')
plt.show()
```

```python
sns.set(rc={"axes.facecolor":"#616161" , "axes.grid" : False})

plt.figure(figsize=(8,7))
sns.boxenplot(x=df["sepal_length"], y=df["species"], palette='spring')
plt.show()
```
