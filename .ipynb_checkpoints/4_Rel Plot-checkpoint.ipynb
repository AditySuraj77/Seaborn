```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
plt.figure(figsize=(7,5))
sns.relplot(x='species',y='sepal_length', data=iris)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.relplot(x='sepal_width',y='sepal_length', data=iris, hue='species', height=5.0)
plt.show()
```
