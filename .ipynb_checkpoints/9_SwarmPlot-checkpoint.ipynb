```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
sns.swarmplot(data=iris, y='species')
plt.show()
```

```python
sns.swarmplot(data=iris, y='species',x='sepal_length')
plt.show()
```

```python
sns.swarmplot(data=iris, y='species',x='sepal_length',hue='species', legend=False)
plt.show()
```

```python
sns.swarmplot(data=iris, y='species',x='sepal_length',hue='species', palette='deep')
plt.show()
```

```python
sns.swarmplot(data=iris, y='species',x='sepal_length',hue='species', palette='deep', dodge=True)
plt.show()
```

```python
sns.swarmplot(data=iris, y='species',x='sepal_length',hue='species', palette='deep', orient='h')
plt.show()
```

```python
sns.swarmplot(data=iris, y='species',x='sepal_length',hue='species', palette='deep',size=10)
plt.show()
```

```python
sns.swarmplot(data=iris, y='species',x='sepal_length',hue='species', palette='deep',size=10, marker="<")
plt.show()
```
