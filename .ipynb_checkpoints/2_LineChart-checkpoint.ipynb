```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset('iris')
df.head()
```

# Line plot

```python
sns.lineplot(x="petal_length", y="sepal_width", data=df )
plt.show()
```

# LineWidth

```python
sns.lineplot(x="petal_length", y="sepal_width", data=df, linewidth = 5.0 )
plt.show()
```

# Color

```python
plt.figure(figsize=(7,5))
sns.lineplot(x="petal_length", y="sepal_width", data=df, color = "g", linewidth= 5)
plt.show()
```

# Hue

```python
plt.figure(figsize=(7,5))
sns.lineplot(x="petal_length", y="sepal_width", data=df, color = "g", linewidth= 5, hue="species")
plt.show()
```

# Style & Marker

```python
plt.figure(figsize=(7,5))
sns.lineplot(x="petal_length", y="sepal_width", data=df, color = "g", linewidth= 5, hue="species", style='species',markers=True)
plt.show()
```

# Plot Figure Size

```python
plt.figure(figsize=(7,5))
sns.lineplot(x="petal_length", y="sepal_width", data=df, color = "g", linewidth= 5, hue="species", style='species',markers=True)
sns.set(style='dark')
plt.show()
```

# Plot Title

```python
plt.figure(figsize=(7,5))
plt.title("Iris ")
sns.lineplot(x="petal_length", y="sepal_width", data=df, color = "g", linewidth= 5, hue="species", style='species',
             markers=True)
plt.show()
```

# Grid

```python
sns.lineplot(x="petal_length", y="sepal_width", data=df )
sns.set_style('whitegrid')
plt.show()
```

# Background Color

```python
sns.set(rc={'axes.facecolor':'lightblue', 'figure.facecolor':'lightgreen'})
sns.lineplot(x="petal_length", y="sepal_width", data=df )
sns.set_style('whitegrid')
plt.show()
```
